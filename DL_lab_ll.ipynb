```python
import torch
```

```python
class DenseLayer:
  def __init__(self,n_features,n_neurons):
    self.weights = 0.01 * torch.rand(n_neurons,n_features)
    self.bias = torch.zeros((1,n_neurons))
  def forward(self, inputs):
    self.output = torch.matmul(inputs,self.weights.T) + self.bias
```

```python
input = torch.rand(32,5)
input, input.size()
```

```python
first_layer = DenseLayer(5,16)
first_layer.forward(input)

first_layer.output
```

```python
second_layer = DenseLayer(16,16)
second_layer.forward(first_layer.output)

second_layer.output
```

```python
third_layer = DenseLayer(16,16)
third_layer.forward(second_layer.output)

third_layer.output
```

```python
output = DenseLayer(16,5)
output.forward(third_layer.output)

output.output
```
